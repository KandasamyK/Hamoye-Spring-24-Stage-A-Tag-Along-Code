```python
#Q.No:5
    
data_subset = df[['Y2017', 'Area']]
area_sum_2017 = data_subset.groupby('Area')['Y2017'].sum()
sorted_sum_2017 = area_sum_2017.sort_values(ascending=True)
seventh_lowest_area = sorted_sum_2017.index[6]
print(f"Area with the 7th lowest sum in 2017: {seventh_lowest_area}")
```

```python
# Q.No.9
processing_rows = df[df['Element'].str.contains('Processing', case=False)] 
processing_2017_sum = processing_rows.groupby('Element')['Y2017'].sum()
total_sum = processing_2017_sum.values[0] if len(processing_2017_sum) > 0 else 0 
print(f"Total sum of Processing in 2017: {total_sum}")
```

```python
# Q.NO.13
unique_countries = df['Area'].unique()
number_of_countries = len(unique_countries) 
print(f"Total number of unique countries: {number_of_countries}")
```

```python
# Q.No.14
missing_values = df['Y2014'].isnull().sum()
total_non_missing = df.shape[0] 
missing_percentage = (missing_values / total_non_missing) * 100
missing_percentage = missing_percentage.round(3) 
print(f"Total number of missing values in 'Y2014': {missing_values}")
print(f"Percentage of missing values in 'Y2014' (rounded to 3 decimals): {missing_percentage:.3f}")
```

```python
#Q.No 15
stock_variation_data = df[df['Element'] == 'Stock Variation']
year_sum_variation = stock_variation_data.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
year_highest_variation = year_sum_variation.idxmax()
print(f"Year with the highest sum of 'Stock Variation': {year_highest_variation}")

```

```python
#Q.No 16
data_subset = df[['Y2017', 'Area']]
area_sum_2017 = data_subset.groupby('Area')['Y2017'].sum()
area_highest_sum = area_sum_2017.idxmax()
print(f"Area with the highest sum in 2017: {area_highest_sum}")
```

```python
# Q.No.18

madagascar_2015 = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)') & (df['Y2015']>0)]
# Calculate the total protein supply quantity in Madagascar for 2015
total_protein_2015 = madagascar_2015['Y2015'].sum()
print("Total Protein Supply Quantity in Madagascar for 2015:", total_protein_2015)
```

```python
# Q.NO.19
data_2017 = df[['Y2017']]
mean_2017 = data_2017.mean().round(2) 
std_dev_2017 = data_2017.std().round(2)
print(f"Mean for year 2017 (rounded to 2 decimals): {mean_2017.values[0]}")
print(f"Standard deviation for year 2017 (rounded to 2 decimals): {std_dev_2017.values[0]}")
```
